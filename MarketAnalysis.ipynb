```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
products = [
    "Bread",
    "Milk",
    "Butter",
    "Jam",
    "Eggs",
    "Tea",
    "Biscuits",
    "Coffee",
    "Sugar",
    "Rice",
    "Oil",
    "Pulses"
]

transactions = []

for i in range(1, 501):

    basket = []

    # Hidden patterns

    if random.random() < 0.7:
        basket.extend(["Bread", "Butter"])

    if random.random() < 0.6:
        basket.extend(["Tea", "Biscuits"])

    if random.random() < 0.6:
        basket.extend(["Rice", "Oil"])

    if random.random() < 0.5:
        basket.extend(["Coffee", "Sugar"])

    extra_items = random.sample(products, random.randint(1, 3))

    basket.extend(extra_items)

    basket = list(set(basket))

    transactions.append([
        i,
        ",".join(basket)
    ])

df = pd.DataFrame(
    transactions,
    columns=[
        "TransactionID",
        "Items"
    ]
)

df.to_csv(
    "dataset.csv",
    index=False
)

print("Dataset Saved Successfully")
print(df.head())
```

```python
df = pd.read_csv("dataset.csv")

df.head()
```

```python
transactions = []

for items in df["Items"]:

    transactions.append(
        items.split(",")
    )

print(transactions[:5])
```

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

basket_df = pd.DataFrame(
    te_array,
    columns=te.columns_
)

basket_df.head()
```

```python
frequent_itemsets = apriori(
    basket_df,
    min_support=0.05,
    use_colnames=True
)

print(frequent_itemsets.head())
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.30
)

rules.head()
```

```python
important_rules = rules[
    [
        "antecedents",
        "consequents",
        "support",
        "confidence",
        "lift"
    ]
]

important_rules.sort_values(
    by="lift",
    ascending=False,
    inplace=True
)

important_rules.head(20)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    rules["support"],
    rules["confidence"]
)

plt.xlabel("Support")

plt.ylabel("Confidence")

plt.title("Support vs Confidence")

plt.show()
```

```python
top_rules = important_rules.head(10)

plt.figure(figsize=(10,5))

plt.bar(
    range(len(top_rules)),
    top_rules["lift"]
)

plt.xticks(
    range(len(top_rules)),
    rotation=45
)

plt.title("Top Association Rules by Lift")

plt.show()
```

```python
important_rules.to_csv(
    "association_rules.csv",
    index=False
)

print("Rules Saved Successfully")
```
